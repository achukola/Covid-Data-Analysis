```python
# Remove punctuations
def no_punc(N):
    no_punc = [w for w in N if w not in '"#$%&\'()*+,-/:;<=>?@[\\]^_`{|}~“”’—']
    no_puncText = ''.join(no_punc)
    return no_puncText
N['no_punc'] = N['text'].apply(lambda x: no_punc(x))
```

```python
ADJ_P = []
for i in range(0,len(sensen)):
    adjsent = word_tokenize(sensen[i])
    tag = nltk.pos_tag(adjsent)
    grammar = RegexpParser("""ADJP: {<RB>+<JJ> | <JJ><IN>+ }""")
    tree = grammar.parse(tag)
    new_tag = tree.pos()
    Temp = []
    for j in range(0, len(new_tag)):
        if new_tag[j][1] == "ADJP":
            Temp.append(new_tag[j][0][0])
    if len(Temp) == 0:
        continue
    else:
        Temp2 = ' '.join(Temp)
        Temp3 = (Temp2,'Adjective Phrase')
        ADJ_P.append(Temp3)
```

```python
#VBD,VB,VBN,VBP,VBZ,VBG.....VB	verb (ask)
#VBG	verb gerund (judging)
#VBD	verb past tense (pleaded)
#VBN	verb past participle (reunified)
#VBP	verb, present tense not 3rd person singular(wrap)
#VBZ	verb, present tense with 3rd person singular (bases)
N = []
V = []
Adjw =[]
for x in pos_tokens:
    if x[1] == 'NN' and (x[0].isalpha() or x[0] == '!'):
        N.append(x[0])
    elif x[1] == 'VBD' and (x[0].isalpha() or x[0] == '!'):
        V.append(x[0])
    elif x[1] == 'JJ' and (x[0].isalpha() or x[0] == '!'):
        Adjw.append(x[0])
```
